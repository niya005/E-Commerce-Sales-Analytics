```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(
    "Amazon Sale Report.csv",
    low_memory=False
)

df.head()
```

```python
df['ship-state'].unique()
```

```python
df['ship-state'] = (
    df['ship-state']
    .str.upper()
    .str.strip()
)
```

```python
df['ship-state'].unique()
```

```python
state_mapping = {
    'ORISSA':'ODISHA',
    'PONDICHERRY':'PUDUCHERRY',
    'NEW DELHI':'DELHI',
    'GOA':'GOA',
    'RAJASTHAN':'RAJASTHAN',
    'PUNJAB/MOHALI/ZIRAKPUR':'PUNJAB',
    'PB':'PUNJAB',
    'AR':'ARUNACHAL PRADESH',
    'APO':'ARUNACHAL PRADESH'
}

df['ship-state'] = df['ship-state'].replace(state_mapping)
```

```python
sorted(df['ship-state'].dropna().unique())
```

```python
df.drop(
    columns=['promotion-ids', 'fulfilled-by', 'Unnamed: 22'],
    inplace=True
)
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
```

```python
df['Month'] = df['Date'].dt.month_name()
```

```python
df['Month'].unique()
```

```python
import numpy as np
```

```python
df['Net_Revenue'] = np.where(
    df['Status'].str.upper() == 'CANCELLED',
    0,
    df['Amount']
)
```

```python
df[['Status','Amount','Net_Revenue']].head(20)
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(
    subset=['ship-city','ship-state','ship-postal-code','ship-country'],
    inplace=True
)
```

```python
df['Amount'] = df['Amount'].fillna(0)
```

```python
df['Net_Revenue'] = df['Net_Revenue'].fillna(0)
```

```python
df['currency'] = df['currency'].fillna('INR')
```

```python
df['Courier Status'] = df['Courier Status'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df.to_csv('clean_amazon_sales.csv', index=False)
```
